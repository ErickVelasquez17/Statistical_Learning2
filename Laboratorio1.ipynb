# **Preprocesamiento de Datos.**

```python
#Eliminamos las columnas innecesarias que serian: PassengerId, Name, Ticket

dataset.drop(['Name','Ticket'], axis = 1, inplace = True)
```

```python
#verificación de NAs
dataset.isnull().sum()
```

```python
# Remover NAs 
dataset['Age'].fillna(dataset['Age'].mean(), inplace = True)
dataset['Cabin'].fillna('N', inplace = True)
dataset['Embarked'].fillna('N', inplace = True)
```

```python
#verificación de NAs
dataset.isnull().sum()
```

```python
dataset['Embarked'].value_counts()
```

```python
dataset['Sex'].value_counts()
```

```python
dataset['Cabin'].value_counts()
```

```python
#One Hot Encoding
ohe = pd.get_dummies(dataset['Sex'], drop_first=True)
dataset.drop('Sex', axis=1, inplace=True)
dataset = dataset.join(ohe)
dataset.head()
```

```python
dataset['Cabin'] = dataset['Cabin'].str[:1]
```

```python
ohe = pd.get_dummies(dataset['Embarked'], drop_first=True)
dataset.drop('Embarked', axis=1, inplace=True)
dataset = dataset.join(ohe)
dataset.head()
```

```python
dataset
```

```python
from sklearn import preprocessing

def encode_features(dataset) :
    features = ['Cabin']
    
    for feature in features :
        le = preprocessing.LabelEncoder()
        le = le.fit(dataset[feature])
        dataset[feature] = le.transform(dataset[feature])
        
    return dataset

dataset = encode_features(dataset)
dataset.head()
```

```python
y = dataset['Survived']
X = dataset.drop('Survived', axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
std_scaler = StandardScaler()
X = std_scaler.fit_transform(X)
```

```python
#partición de data para entrenamiento y prueba.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

**Descripción de Arquitectura de Red con Keras**

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
inputs = X_train.shape[1]

#base secuencial para definir la estructura de la Red (SIEMPRE INICIAR CON SQUENTIAL)
clasificador = Sequential()

#primera capa oclta
clasificador.add(Dense(input_dim = inputs, units=6, activation='relu', kernel_initializer='uniform'))

#segunda capa oculta
clasificador.add(Dense(units=6, activation='relu', kernel_initializer='uniform'))

#capa de salida
clasificador.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
```

```python
#parametros de optimización
clasificador.compile(optimizer='Nadam', loss='binary_crossentropy', metrics=['accuracy'])

#parametros de entrenamiento.
clasificador.fit(X_train, y_train, batch_size=25, epochs=150)
```

```python
# Guardardamos la red.
path = '/content/drive/MyDrive/Statistical Learning II/' 
clasificador.save(path + '6_6_sig_out.h5')
```

```python
# Cargamos la red.
new_model = keras.models.load_model(path + '6_6_sig_out.h5')
```

```python
#predicciones
y_preds = new_model.predict(X_test)
y_preds
```

```python
y_preds = (y_preds >= 0.5)
y_preds
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_preds)
```

```python
#KFolds
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
def red_v1():
  #base secuencial para definir la estructura de la Red (SIEMPRE INICIAR CON SQUENTIAL)
  clasificador = Sequential()

  #primera capa oclta
  clasificador.add(Dense(input_dim = 11, units=8, activation='relu', kernel_initializer='TruncatedNormal'))

  #segunda capa oculta
  clasificador.add(Dense(units=8, activation='relu', kernel_initializer='TruncatedNormal'))

  #capa de salida
  clasificador.add(Dense(units=1, activation='sigmoid', kernel_initializer='TruncatedNormal'))

  #parametros de optimización
  clasificador.compile(optimizer='SGD', loss='binary_crossentropy', metrics=['accuracy'])

  return clasificador
```

```python
clasificador = KerasClassifier(build_fn=red_v1, batch_size=25, nb_epoch=1000)
```

```python
kfolds = KFold(n_splits=10, shuffle=True)

accs = cross_val_score(clasificador, X=X_train, y=y_train, cv=kfolds, verbose=1, n_jobs=-1)
```
